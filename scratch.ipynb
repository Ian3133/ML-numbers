```python
# Mushrooms data set
file_path = "data/agaricus-lepiota.data"
cols = ["cap-shape", "cap-surface", "cap-color", "bruises", "odor", "gill attachment", "gill spacing", "gill size", "gill color", "stalk surface above ring", "stalk surface below ring", "stalk color above", "stalk color below", "veil type", "veil color", "ring number", "ring type", "spore print color", "population", "habiat" ]
muhsrooms = pd.read_csv(file_path, names=cols)
muhsrooms.head()
```

```python
tokenizer = tf.keras.preprocessing.text.Tokenizer()
tokenizer.fit_on_texts(muhsrooms)

# Tokenize the dataset
tokenized_dataset = tokenizer.texts_to_sequences(muhsrooms)

# Print the tokenized dataset
#for instance in tokenized_dataset:
#    print(instance)
```

tutoral dataset

```python
file_path = "data/magic04.data"
cols = ["f-height", "fWidth", "fsize", "fConc", "fconc1", "fAsym", "Fm3Long", "fM3Trans", "fAlpha", "fDist", "class"]
data = pd.read_csv(file_path, names=cols)
data.head()
```

```python
data["class"] = (data["class"] =='g').astype(int)
#data.head()
```

```python
#see how each of the col effect the data 
for label in cols[:-1]:
    plt.hist(data[data["class"] == 1 ][label], color="black", label="gamma", alpha=.07, density=True)
    plt.hist(data[data["class"] == 0 ][label], color="red", label="hadron", alpha=.07, density=True)
    plt.title(label)
    plt.ylabel("prob")
    plt.xlabel(label)
    plt.legend()
    plt.show()
```
